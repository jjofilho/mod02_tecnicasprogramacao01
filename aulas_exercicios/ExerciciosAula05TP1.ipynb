1) Utilizando o DataFrame importado anteriormente (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
df_notas = pd.read_csv('/content/drive/My Drive/alunos3.csv', sep=';')
df_notas
```

```python
# investigando os dados do df, pois ocorreram erros nas atividades seguintes
df_notas.dtypes
```

```python
# tratando os dados do df
colunas = ['Prova_1', 'Prova_2', 'Prova_3']
df_notas[colunas] = df_notas[colunas].apply(lambda x: x.str.replace(',', '.').astype(float))
df_notas
```

```python
df_notas['Prova_4'] = df_notas['Prova_4'].astype(float)
df_notas
```

```python
# coluna da média das notas
df_notas['Media_provas'] = df_notas[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df_notas
```

```python
df_notas.drop('Media', axis=1)
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_media = df_notas['Media_provas'].max()
menor_media = df_notas['Media_provas'].min()
print(f'Maior Média: {maior_media} \nMenor Média: {menor_media}')
```

```python
aluno_maior_media = df_notas.loc[df_notas['Media_provas'].idxmax()]
aluno_menor_media = df_notas.loc[df_notas['Media_provas'].idxmin()]

print(f'Aluno com a maior média: {aluno_maior_media}')
print(f'\nAluno com a menor média: {aluno_menor_media}')
```
